```python
from meteoalarm import MeteoAlarm
```

```python
# Get warnings for multiple countries
warnings = MeteoAlarm(['norway', 'italy'])
```

```python
len(warnings)
```

```python
warnings[0]
```

```python
# Process all warnings
for warning in warnings:
    #print(f"\nWarning for {warning.country}:")
    #print(f"- {warning.headline}")
    #print(f"  {warning.get_event()}")
    print(f"  Area: {warning.severity}")
    #print(f"  Geometry: {warning.geometry}")
    #print(f"  Valid until: {warning.expires}")
```

```python
# Access headlines in different languages
for warning in warnings:
    print(f"\nWarning for {warning.country}:")
    print("Headlines:")
    for lang, event in warning.event.items():
        print(f"  {lang}: {event}")
    for lang, headline in warning.headline.items():
        print(f"  {lang}: {headline}")
    print("Descriptions:")
    for lang, desc in warning.description.items():
        print(f"  {lang}: {desc}")
```
